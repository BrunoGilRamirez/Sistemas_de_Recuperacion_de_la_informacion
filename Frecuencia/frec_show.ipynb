# Se obtiene el valor de las veces que ha aparecio en los documentos a analizar
En una lista se almanena la lista de palabras en cada uno de los documentos a analizar, y en loop for se cuenta por cada palabra en cuantos documentos se encuentran, ese valor se almacena en un diccionario que tiene el siguiente ***palabra:#docConElTermino***

```python
setofndoct= {}
for word,score in list(myrepository.dic.items())[0:30]:
    ndoct=0
    for name,dset in list(myrepository.recordsets.items()):
        if (word in dset):
            ndoct=ndoct+1
            setofndoct[word]=ndoct
        
print (setofndoct)
```

# se declara la funcion de idf
Esta funcion calcula la frecuencia inversa del documento, con la siguiente formula:



el valor de la base es variable, entonces por default esta en 10

```python
def idf(ndoct,ndoc,base=10):
    return (math.log((ndoc/ndoct), base) +1)
```

# Se obtiene el idf para cada palabra
Con un loop for en se va almacenando el valor de idf calculado para cada palabra en un diccionario.

```python
setofIdf={}
ndoc=2
for element,times in list(setofndoct.items()):
    setofIdf[element]= idf(times,ndoc,10)
print(setofIdf)
```

# Se Obtiene la norma
Con la formula:


Se calcula la sumatoria con un loop for, que ubica el valor de idf en el setofIdf para cada palabra, elevandola al 2, luego ese se multiplica por el valor de la frecuencia de la palabra. 

```python
Sumt=0
sett=myrepository.recordsets["2591-0.txt"]
for word,idft in list(setofIdf.items()):
    if word in sett:
        Sumt=Sumt+((sett.get(word)**2)*(idft**2))
Sumt= math.sqrt(Sumt)

print(Sumt)
```
