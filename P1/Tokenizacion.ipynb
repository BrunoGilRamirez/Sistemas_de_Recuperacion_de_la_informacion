### La forma tediosa y lenta de Tokenizar un texto
1. Se necesitan un data structure como una lista de python, un aux como **word** y un str con los espacios y signos de puntuacion que se deben excluir de la tokenizacion.
2. Se debe recorrer el arreglo caracter por caracter, y para no petar la pila del SO se pone el tope hasta 100 el elementos.
4. Si el caracter no está en **spaces**, entonces se concatena el caracter a los anteriores. Para evitar tokenizar signos y espacios.
5. Si se encuentra con algun espacio o signo en **spaces**, se usa **append** para agregar la palabra en word a la lista **dat**.

```python
word=""
spaces=" ,;<>|""''«».\n\n*"
dat= []
for indice in range(100):
    if (not(text[indice] in spaces)):
        word=word+text[indice] 
    if(text[indice] in spaces):
        dat.append(word)
        print(dat)
        word=""
```

### La Forma Facil y mas rapida
Los objetos tipo str tienen un metodo que devuelve una lista con los tokens de un texto, el metodo **split()** ejecuta el trabajo de tokenizar el texto entero en **.4s** mientras que hacerlo de la forma mas tradicional tomaria mas de 1 segundo entero. 

```python
a=text.split()
print(len(a),type(a),a)
```
